# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing as HWES
from time import time
```

## Autoregressive Model (AR)

The AR model stands for autoregressive model, meaning that predictions are done by creating a regression. The auto refers to how a prediction is made for something based on past values of that something.

![milk](https://chito.ge/7AMC9EA.png 'Milk Sales Per Year')

The above figure shows an example plot of milk sales per year, with the months dotted out as data points. The figure reveals a clear, seasonal pattern. Assume the following:

$
\begin{bmatrix}
  M_t = \text{Milk this month}
\\M_\text{t-1} = \text{Milk last month}
\\M_\text{t-12} = \text{Milk last year}
\end{bmatrix}
$

The value to be predicted here is $M_t$, and the values that are available for this prediction are $M_\text{t-1}$ to $M_\text{t-12}$, and however many much more data there is. A na√Øve approach would be to use all the data, with the idea that the prediction should be accurate due to there being no missing data. The result may be good, but this approach is prone to many statistical issues such as overfitting. In statistics, if a simpler model can do the job, then that model is preferred.

Thus, it's necessary to find out which lags are good starting points to use for the model, which is done by looking at the PACF of the data.

```python
def parser(s):
    return datetime.strptime(s, '%Y-%m-%d')
```

```python
ice = pd.read_csv('data/ice_cream.csv', parse_dates=[0], index_col=0, date_parser=parser)
ice.rename(columns={'IPN31152N': 'production'}, inplace=True)
ice = ice.asfreq(pd.infer_freq(ice.index))
ice = ice[pd.to_datetime('2010-01-01'):]
```

```python
plt.figure(figsize=(10,5))
plt.plot(ice)
plt.title('Ice Cream Production over Time', fontsize=14)
plt.ylabel('Production', fontsize=14)
for year in range(2010, 2021):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
plt.show()
```

```python
iceacf = plot_acf(ice, lags=100)
```

```python
iceacf = plot_pacf(ice, method='ywm')
```

The ACF plot shows that the autocorrelation is diminishing over time, and the PACF plot shows that there are several lags that could potentially be used for an autoregressive model. Based on this data, an example autoregressive model for the prediction of $M_t$ could look like:

$
M_t = 
\beta_0 + 
\beta_1i_\text{t-1} + 
\beta_2i_\text{t-2} + 
\beta_3i_\text{t-3} + 
\beta_7i_\text{t-7} +
\beta_\text{10}i_\text{t-10} + 
\beta_\text{13}i_\text{t-13} + 
\epsilon_t
$

For the fitting of the model, the up-to-date SARIMAX model will be used, allowing for freedom of inclusion or exclusion over each of the components (such as AR or MA) that the model is composed of.

```python
train_end = datetime(2018,12,1)
test_end = datetime(2019,12,1)

ice_train = ice[:train_end]
ice_test = ice[train_end + timedelta(days=1):test_end]
```

```python
#Three numbers in order are equal to AR, I, MA
model = SARIMAX(ice_train, order=([1, 2, 3, 7, 10, 13],0,0)) #, seasonal_order=(1, 0, 0, 12)
start = time()
model_fit = model.fit()
end = time()
print(f"Model fitting time: {end - start}")
```

```python
model_fit.summary()
```

```python
predictions = model_fit.predict(start=ice_test.index[0], end=ice_test.index[-1])
residuals = ice_test['production'] - predictions
```

```python
plt.figure(figsize=(10,5))
plt.plot(ice_test)
plt.plot(predictions)
plt.legend(('Data', 'Predictions'), fontsize=14)
plt.title('Ice Cream Production over Time', fontsize=14)
plt.ylabel('Production', fontsize=14)
for year in range(2019, 2021):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
plt.show()
```

```python
# Average percent error between data and predictions
print(f"Mean Absolute Percent Error: {round(np.mean(abs(residuals/ice_test['production'])),4)}")
```

```python
# The standard deviation of the residuals
print(f"Root Mean Squared Error: {np.sqrt(np.mean(residuals**2))}")
```

## Moving Average Model (MA)

The MA model stands for a Moving Average model which is a time series model that accounts for very short-run autocorrelation. It basically states that the next observation is the mean of every past observation.

An example can be give about bringing cupcakes to a party hosted by a professor. Note that many of these values are assumptions. The professor is known to be a bit crazy and tends to say that you brought the wrong number of cupcakes by some amount. Assume that on average you always bring 10 cupcakes, thus $\mu = 10$. The number that the professor says you're off by is the $\epsilon_t$. This error is normally distributed and has a standard deviation of 1 in this example: ($N(\mu_\epsilon, \sigma_\epsilon)$), where $\mu_\epsilon$ is 0, and $\sigma_\epsilon$ is 1.

The predicted number of cupcakes to bring can be written as such:

$\hat{F_t} = \mu + \phi_1\epsilon_\text{t-1}$

Where $\mu$ is 10, $\phi_1$ is 0.5, and $\epsilon_\text{t-1}$ is the error from the previous month. This model says that you will always bring 10 cupcakes to the party, but you will adjust that value by 50% of last month's error.

To reiterate, the 4 parameters of this MA model are:
- $\mu$, which is the baseline number of cupcakes
- $\phi_1$, which is the multiplying factor of previous month's error
- $\mu_\epsilon$, which is the mean of the error
- $\sigma_\epsilon$, which is the standard deviation of the error

Example results can be shown as in the following dataframe:

```python
pd.DataFrame({'$$\hat{F_t}$$': [10, 9, 10.5, 10, 11], '$$\epsilon_t$$': [-2, 1, 0, 2, 1], '$$F_t$$': [8, 10, 10.5, 12, 12]})
```

Here, $F_t$ is the number of cupcakes you actually needed to bring. The first month, you brought 10 cupcakes as this is the baseline. However, the professor stated you brought 2 cupcakes too many, making the actual necessary number 8 cupcakes. The next month, the 10 cupcakes ($\hat{F_t}$) are added by 0.5 ($\phi_1$) times the error of last month, being -2 ($\epsilon_\text{t-1}$). This means that 9 cupcakes are brought, as half of -2 is -1, which is subtracted from 10. However, the professor expected an additional cupcake, making this month's error 1. This process continues for the other months into the future.

This example moving average model is the simplest type and is referred to as an MA1 model, as only one error term was taken into account. This means that for $\hat{F_t}$, only information from last month's error was included ($\epsilon_\text{t-1}$). 

An example MA2 model would look like:

$\hat{F_t} = \mu + \phi_1\epsilon_\text{t-1} + \phi_2\epsilon_\text{t-2}$

And the actual cupcakes you would need to bring would look like:

$F_t = \mu + \phi_1\epsilon_\text{t-1} + \phi_2\epsilon_\text{t-2} + \epsilon_t$

This means that the first formula refers to the predicted number of cupcakes to bring, and the second formula is the actual number of cupcakes to bring. This difference can also be observed the created dataframe.

The following code example will follow these formulas:

## $Y_t = 50 + 0.4\epsilon_\text{t-1} + 0.3\epsilon_\text{t-2} + \epsilon_t$

## $\epsilon_t = N(0, 1)$

```python
errors = np.random.normal(0, 1, 400)
```

```python
date_index = pd.date_range(start='9/1/2019', end='1/1/2020')
```

```python
mu = 50
series = []
for t in range(1, len(date_index)+1):
    series.append(mu + 0.4*errors[t-1] + 0.3*errors[t-2] + errors[t])
```

```python
series = pd.Series(series, date_index)
series = series.asfreq(pd.infer_freq(series.index))
```

```python
plt.figure(figsize=(10,5))
plt.plot(series)
plt.axhline(mu, linestyle='--', color='grey')
plt.xticks(rotation=25, ha="right")
plt.show()
```

```python
maacf = plot_acf(series)
```

```python
mapacf = plot_pacf(series, method='ywm')
```

It can be observed from the ACF plot that the autocorrelations quickly die down and do not show an oscillating pattern like in the autoregressive model's example. This is a strong indication that the data is suitable for a moving average model instead. The result suggest that maybe a MA1 or MA2 model would be suitable to use. The PACF plot has its characteristics of alternating and maybe diminishing over time, which is more evidence that a moving average process is at play here. LOOK AT ACF PLOT TO DETERMINE ORDER FOR MA PART.

```python
train_end = datetime(2019,12,30)
test_end = datetime(2020,1,1)

ma_train = series[:train_end]
ma_test = series[train_end + timedelta(days=1):test_end]
```

```python
#Three numbers in order are equal to AR, I, MA
model = SARIMAX(ma_train, order=(0,0,2), trend='c')
start = time()
model_fit = model.fit()
end = time()
print(f"Model fitting time: {end - start}")
```

It's important to note here that after a number of lags has been chosen for the moving average, it's only possible to predict that many periods into the future. Afterwards, the prediction will simply predict the constant mean (50 in this case). This is why the difference in days between the training and testing data in this example is only 2 days.

```python
model_fit.summary()
```

Thus, the prediction model can be written as such:

## $Y_t = 50 + ma.L1\epsilon_\text{t-1} + ma.L2\epsilon_\text{t-2}$
The values for ma.L1 and ma.L2 can be found in the SARIMAX summary, in the coef column. Note that these values will change every time the Notebook's code is run.

```python
pd.read_html(model_fit.summary().tables[1].as_html(),index_col=0,header=0)[0]
```

```python
predictions = model_fit.predict(start=ma_test.index[0], end=ma_test.index[-1])
residuals = ma_test - predictions
```

```python
print(predictions)
```

```python
plt.figure(figsize=(10,5))
plt.plot(series[-14:])
plt.plot(predictions)
plt.legend(('Data', 'Predictions'), fontsize=16)
plt.xticks(rotation=25, ha="right")
plt.show()
```

```python
# Average percent error between data and predictions
print(f"Mean Absolute Percent Error: {round(np.mean(abs(residuals/ma_test)),4)}")
```

```python
# The standard deviation of the residuals
print(f"Root Mean Squared Error: {np.sqrt(np.mean(residuals**2))}")
```

## Autoregressive Moving Average Model (ARMA)

It's possible to combine the autoregressive and moving average models, and it's possible that both might be necessary at the same time. An example can be given to illustrate what kind of model may be necessary when spotting a time series in the wild. In this example, a lightbulb manufacturer is attempting to figure out how many light bulbs need to be produced every month to meet demands. An ARMA(1,1) model will be used, where each 1 represent the order of the AR and MA parts.

Assume that this volume of lightbulbs to be predicted is $L_t$. The formula of this model would then be:

$L_t = \beta_0 + \beta_1L_\text{t-1} + \mu + \phi_1\epsilon_\text{t-1} + \epsilon_t$

This formula predicts the number of lightbulbs to produce this month ($L_t$). The ($\beta_0$) is a coefficient that's constant, which is added to another coefficient ($\beta_1$) multiplied by the number of lightbulbs that was needed to be produced last month ($L_\text{t-1}$). In addition to this AR part, the MA part adds a coefficient ($\phi_1$) multiplied by the error from the previous time period ($\epsilon_\text{t-1}$), and plus the error of this month ($\epsilon_t$).

The formula for the actual values would then be:

$\hat{L_t} = \beta_0 + \beta_1L_\text{t-1} + \mu + \phi_1\epsilon_\text{t-1}$

There is no access to the errors of this month for the actual values as the event has not happened yet. If this were accessible, perfect predictions would always be made. Also, the $\mu$ of the MA model can also be incorporated into the $\beta_0$ in this case.

In order to figure out which orders to use for the AR and MA bits, the ACF and PACF plots can once again be used. The PACF measures direct effects, which is what the autoregressive part of the model would do, and the ACF measures the moving average part of the model. There are many nuances and exceptions, but this serves as good rule of thumb.

```python
def parser(s):
    return datetime.strptime(s, '%Y-%m-%d')
```

```python
catfish = pd.read_csv("data/catfish.csv", parse_dates=[0], index_col=0, date_parser=parser)
catfish = catfish.asfreq(pd.infer_freq(catfish.index))
```

```python
start_date = datetime(2000,1,1)
end_date = datetime(2004,1,1)
catfish = catfish[start_date:end_date]
```

```python
plt.figure(figsize=(10,5))
plt.plot(catfish)
plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Sales', fontsize=14)
for year in range(2000, 2005):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
plt.axhline(catfish['Total'].mean(), color='red', linestyle='--', alpha=0.2)
plt.show()
```

```python
first_diff = catfish.diff()[1:]
```

The diff() function calculates the difference of a dataframe element compared with another element in the Dataframe (default is element in previous row). This is used in an attempt to enforce stationarity in this example as the original catfish data appears to have an upward trend.

```python
plt.figure(figsize=(10,5))
plt.plot(first_diff)
plt.title('First Difference of Catfish Sales', fontsize=14)
plt.ylabel('Sales', fontsize=14)
for year in range(2000, 2005):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
plt.axhline(first_diff['Total'].mean(), color='red', linestyle='--', alpha=0.2)
plt.show()
```

```python
catacf = plot_acf(first_diff, lags=20)
```

```python
catpacf = plot_pacf(first_diff, method='ywm', lags=20)
```

Based on the ACF plot, an MA1 order can first be tested as it's relatively strong, and the process shuts off in the second lag. Based on the PACF plot, it's best to start with an AR4 order.

```python
train_end = datetime(2003,7,1)
test_end = datetime(2004,1,1)

catfish_train = first_diff[:train_end]
catfish_test = first_diff[train_end + timedelta(days=1):test_end]
```

```python
model = SARIMAX(catfish_train, order=(4,0,1), trend='c')
start = time()
model_fit = model.fit()
end = time()
print(f"Model fitting time: {end - start}")
```

```python
model_fit.summary()
```

The summary shows how the value of the time series of the current period is negatively correlated with the lagged versions of itself. The moving average is positive. There is also a constant, shown here as the intercept, which has a very high P value, indicating that it is statistically insignificant. The P values generally need to be lower than 0.05 to be statistically significant. In this case, all the AR lags suffice, and the MA lag is a bit higher than 0.05. First tutorial purposes, I'm keeping the MA.

So the ARMA(4,1) model would look like this:

$\hat{Y}_t = -0.8y_\text{t-1} -0.45y_\text{t-2} -0.58y_\text{t-3} -0.61y_\text{t-4} + 0.43\epsilon_\text{t-1}$

```python
predictions = model_fit.predict(start=catfish_test.index[0], end=catfish_test.index[-1])
residuals = catfish_test['Total'] - predictions
```

```python
plt.figure(figsize=(10,5))
plt.plot(catfish_test)
plt.plot(predictions)
plt.legend(('Data', 'Predictions'), fontsize=14)
plt.title('First Difference of Catfish Sales', fontsize=14)
plt.ylabel('Production', fontsize=14)
plt.show()
```

```python
# Average percent error between data and predictions
print(f"Mean Absolute Percent Error: {round(np.mean(abs(residuals/catfish_test['Total'])),4)}")
```

```python
# The standard deviation of the residuals
print(f"Root Mean Squared Error: {np.sqrt(np.mean(residuals**2))}")
```

These results have their pros and cons, as the shape of the data is captured, but there is a large root mean squared error to account for. This, however, is only an ARMA implementation and the results can be improved with the upcoming SARIMA implementation which is able to capture seasonality.

## Autoregressive Integrated Moving Average Model (ARIMA)

The ARIMA model incorporates another letter into the previously explored ARMA model. Its purpose can be further explored through the usage of an example. This example is about an anchor salesman that attempts to sell a number of anchors every month, referred to as $A_t$. The data of these anchor sales are visible in the following image.

![Anchor Sales](https://chito.ge/3oC6udQ.png 'Anchor Sales')

The X axis contains time in months, and the Y axis contains anchor sales in 1000s. The figure also shows a clear trend. Because of this, it's not possible to use an AR, MA, or ARMA model due to the data not being stationary. There is no seasonality, a constant variance, but the mean does not remain constant over time. This is where the ARIMA model is used, which is utilized when the data seems stationary except for a mean that does not remain constant. The I stands for integrated.

In this context, integrated means that instead of predicting the time series itself, it will predict differences in the time series between one timestamp and the previous timestamp. This means that a new time series is created through a transformation. In this scenario, this is done through the following formula:

$Z_t = A_\text{t+1} - A_t$

This formula is taking, for example, the number of anchors solf in February ($A_\text{t+1}$) and minus the number of anchors sold in the previous month ($A_t$). The result is visible in the following image.

![Differences Between Timestamps](https://chito.ge/53wd8Rd.png 'Differences Between Timestamps')

The original dataset seems to be a linear trend, with some noises added to it. In a linear function, to get from one point on the X axis to the next, you add a constant value. Thus, when you take the differences between timestamps in a linear function, you always return some constant values. Similar behaviour is expected of this dataset. Even with taking the error taken into account, the results are hovering around some constant value, visible in the above figure that's plotting the $Z_t$ values over time.

This is basically what an ARIMA model is. Where the ARMA model has two parameters of orders, the ARIMA model has three, which are referred to as p, d, and q. The p and q parameters are the same, with p referring to the order of the AR part, and q referring to the order of the MA part. The d refers to the order of the integrated part of the model. In the example a d of 1 was chosen as only a difference of 1 ($A_\text{t+1}$ and $A_t$). However, it's possible to make things more complicated. This can be done by making a second difference which is done by taking the $Z_t$ and transforming it again. This means creating a new series such as $W_t$ which would equal $Z_\text{t+1} - Z_t$. It's also possible to go further and create a third, and so on. Usually it suffices to stick to one difference ($Z_t$ in this example).

Thus, the simplest form of ARIMA is ARIMA(1,1,1), which is written as:

$Z_t = \phi_1Z_\text{t-1} + \theta_1\epsilon_\text{t-1} + \epsilon_t$

Here, $Z_t$ is what we want to predict, $\phi_1Z_\text{t-1}$ is the autoregressive part, $\theta_1\epsilon_\text{t-1}$ is the moving average part with theta being the coefficient, and $\epsilon_t$ is the usual error term.

Now the question is, how do you get back from $Z_t$ to $A_t$, as you don't want to predict the differences, you want to predict the actual values? Assuming that you want to recover the value for $A_t$, it is equal to $Z_\text{t-1} + A_\text{t-1}$, as this states that the predicted value of a certain time period, is equal to the difference between that time period and the previous time period. This also holds true for $A_t = Z_\text{t-1} + Z_\text{t-2} + A_\text{t-2}$, and so on. This is illustrated in the following figure.

![Transforming Zt back to At](https://chito.ge/9f1s2mM.png 'Transforming Zt back to At')

## Seasonal Autoregressive Integrated Moving Average Model (SARIMA)

The previous ARIMA model was autoregressive (AR), meaning that it wants to predict values of today based on past values. It was also integrated (I), meaning that a certain trend was available getting rid of it through distancing to enforce stationarity. It was also using a moving average (MA), meaning that errors from previous periods were used to inform the prediction of the time series today.

The added S to the ARIMA model stands for seasonality. Seasonality is a repeating pattern within a year, which happens over and over again over time. The following figure holds the example for this scenario.

![Sales of Donuts Per 3 Months](https://chito.ge/4JyDzdZ.png 'Sales of Donuts Per 3 Months')

The data contains information about donut sales every 3 months. There is a clear seasonal pattern visible in the shape of a W that repeats itself every year. Now how do you know if you should use a SARIMA model? In this case, there is obvious seasonality which satisfies the S part. There might also be AR and MA components as the value of some period could depend on some value of 4 periods ago. There is also a clear upward trend, which requires the integrated (I) part.

A SARIMA model takes 7 parameters, making the model look like: ARIMA(p,d,q)(P,D,Q)m

The lowercase p, d, and q refers to the orders of the AR, I, and MA parts as seen before. The lowercase m refers to the seasonal factor, which is the number of periods within a year it takes for the seasonality to repeat. In this example the seasonal factor is 4. The uppercase P, D, and Q are analogues of the lowercase versions except for the seasonal components. The (p,d,q) part of the model essentially has a 1 written behind it, stating that those operations are performed only once, compared to the factor of 4 for the (P,D,Q) part of the model.

If an ARIMA(1,1,1)(1,1,1)4 model were to be used, its formula would take on the following appearance:

$(1-\phi_1B)(1-\Phi_1B^4)(1-B)(1-B^4)Y_t = (1+\theta_1B)(1+\Theta_1B^4)\epsilon_t$

The $B$ represents the backshift operator (also known as the lag operator $L$), which is a variable but shifted back one time period. An example is $LY_t$ which is the same as $Y_\text{t-1}$.

The $(1-\phi_1B)$ corresponds to 'p', saying to take into account the time series one period ago for the predictions.<br>
The $(1-\Phi_1B^4)$ corresponds to 'P', saying to take into account the time series four periods ago for the predictions.<br>
The $(1-B)$ corresponds to 'd', saying to backshift the whole time series by 1 period.<br>
The $(1-B^4)$ corresponds to 'D', saying to also backshift the time series by 4 periods.<br>
The $(1+\theta_1B)$ corresponds to 'q', saying to take into account error terms from 1 period ago.<br>
The $(1+\Theta_1B^4)$ corresponds to 'Q', saying to take into account error terms from 4 periods ago.

A simpler example can be used to show how these terms can be expanded to their original forms. Here an ARIMA(1,0,0)(0,1,1)4 model will be used. This model's formula is:

$(1-\phi_1B)(1-B^4)Y_t = (1+\Theta_1B^4)\epsilon_t$

This model and formula are saying that predictions need to be made for the time series $Y_t$ by taking into account the time series one period ago, represented by the $(1-\phi_1B)$ part. You also want to take the difference of 4 periods ago, represented by $(1-B^4)$. Finally, a seasonal moving average from 4 periods ago is also taken into account, represented by $(1+\Theta_1B^4)$ for the error.

Simplified by one term turns it into:

$(1-\phi_1B - B^4+\phi_1B^5)Y_t = \epsilon_t + \Theta_1\epsilon_t-4$

Here, the polynomials of $(1-\phi_1B)$ and $(1-B^4)Y_t$ have been multiplied together. The $\epsilon_t$ multiplied by 1 is just itself. The error backshifted 4 times in the past is equal to $\epsilon_t-4$.

Further simplified turns it into:

$Y_t + Y_\text{t-4} = \phi_1Y_\text{t-1} - \phi_1Y_\text{t-5} + \Theta_1\epsilon_\text{t-4} + \epsilon_t$

Here, $Y_t$ is used against all of the other terms on the left side of the equals sign and refactoring has taken place.

Let $Z_t = Y_t - Y_\text{t-4}$

Here, $Z_t$ is a new time series which is defined as the original time series at some point minus that time series 4 points ago. This makes the time series more stationary.

$Z_t = \phi_1Z_\text{t-1} + \Theta_1\epsilon_\text{t-4} + \epsilon_t$

```python
predictions = model_fit.forecast(len(catfish_test))
predictions = pd.Series(predictions, index=catfish_test.index)
residuals = catfish_test['Total'] - predictions
```

```python
plt.figure(figsize=(10,5))
plt.plot(catfish)
plt.plot(predictions)
plt.legend(('Data', 'Predictions'), fontsize=14)
plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Production', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Residuals from SARIMA Model', fontsize=14)
plt.show()
```

These residuals show a systematic bias in the direction of the predictions, due to all of them being below 0. This also observed when plotted against the actual data. This is also the case due to the prediction being very far into the future, making it more innacurate over time.

```python
print(f"Mean Absolute Percent Error: {round(np.mean(abs(residuals/catfish_test['Total'])),4)}")
```

```python
print(f"Root Mean Squared Error: {np.sqrt(np.mean(residuals**2))}")
```

### ROLLING PREDICTION:

```python
rolling_predictions = catfish_test.copy()
for train_end in catfish_test.index:
    catfish_train = catfish[:train_end-timedelta(days=1)]
    model = SARIMAX(catfish_train, order=(0,1,0), seasonal_order=(1,0,1,12))
    model_fit = model.fit()

    pred = model_fit.forecast()
    rolling_predictions['Total'][train_end] = pred
```

```python
rolling_residuals = catfish_test['Total'] - rolling_predictions['Total']
```

```python
plt.figure(figsize=(10,5))
plt.plot(rolling_residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Rolling Residuals from SARIMA Model', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(catfish)
plt.plot(rolling_predictions)
plt.legend(('Data', 'Predictions'), fontsize=14)
plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Production', fontsize=14)
plt.show()
```

```python
print(f"Mean Absolute Percent Error: {round(np.mean(abs(rolling_residuals/catfish_test['Total'])),4)}")
```

```python
print(f"Root Mean Squared Error: {np.sqrt(np.mean(rolling_residuals**2))}")
```
